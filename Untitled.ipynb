# trying example on scipy page to get a better feel

```python
#from scipy import signal
sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
sig_noise = sig + np.random.randn(len(sig))
corr = signal.correlate(sig_noise, np.ones(128), mode='same') / 128
```

```python

clock = np.arange(64, len(sig), 128)
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.plot(clock, sig[clock], 'ro')
ax_orig.set_title('Original signal')

ax_noise.plot(sig_noise)
ax_noise.set_title('Signal with noise')

ax_corr.plot(corr)
ax_corr.plot(clock, corr[clock], 'ro')
ax_corr.axhline(0.5, ls=':')
ax_corr.set_title('Cross-correlated with rectangular pulse')
ax_orig.margins(0, 0.1)
fig.tight_layout()
fig.show()
```

## using numpy  correlate

```python
#using np.correlate...simialr idea to scipy tho
some_data = np.random.uniform(0,1,size=100)
subset = some_data[42:50]

#WITH normalization
mean = np.mean(some_data)
some_data_normalised = some_data - mean
subset_normalised = subset - mean

correlated = np.correlate(some_data_normalised, subset_normalised)
max_index = np.argmax(correlated) #42 is the peak and were data starts to overlap/correlate
plt.plot(correlated)
```

```python
#WITHOUT normalization
correlated2 = np.correlate(some_data, subset)
max_index2 = np.argmax(correlated2) #70 is the peak and were data starts to overlap/correlate
plt.plot(correlated2)
max_index2
```
